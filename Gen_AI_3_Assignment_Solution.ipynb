# Question 1

**What is a Vector Database (VectorDB) and how is it different from traditional databases?**

**Answer:**

VectorDB stores vector embeddings and performs similarity search, while traditional databases store structured records and use exact-match or relational queries.

# Question 2

**Answer:**

Examples include ChromaDB (local AI apps), FAISS (fast similarity search), Pinecone (managed cloud), Weaviate (hybrid search), and Milvus (large-scale AI workloads).

# Question 6

**Install and set up Chroma DB, and insert sample vector data for semantic search.**

**Answer:**

```python
import chromadb
client=chromadb.Client()
col=client.create_collection("docs")
col.add(ids=["1"],documents=["AI is transforming healthcare"])
print(col.query(query_texts=["healthcare AI"],n_results=1))
```
Sample Output:
Returns the most similar stored document.
