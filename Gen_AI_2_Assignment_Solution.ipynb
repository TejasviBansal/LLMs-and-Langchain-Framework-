# Question 1

**What are Large Language Models (LLMs) and how do they function?**

**Answer:**

Large Language Models are deep learning models trained on massive text datasets to predict the next token. They use the Transformer architecture and attention mechanism to understand context and generate human-like text.

# Question 2

**Discuss the impact of LLMs on traditional software development approaches.**

**Answer:**

LLMs speed up coding, testing, documentation, and debugging. They improve developer productivity but require human review for correctness, security, and reliability.

# Question 3

**What are the key advantages and limitations of using LLMs in real-world applications?**

**Answer:**

# Question 6

**Implement a basic LangChain pipeline using OpenAI’s LLM to answer questions based on a user input prompt.**

**Answer:**

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate

llm=ChatOpenAI(model="gpt-4o-mini")
prompt=ChatPromptTemplate.from_template("Answer: {q}")
chain=prompt|llm
print(chain.invoke({"q":"What is AI?"}).content)
```
Sample Output:
AI is the simulation of human intelligence by machines.

# Question 8

**Create a LlamaIndex implementation that indexes a local text file and retrieves answers from it.**

**Answer:**

```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
docs=SimpleDirectoryReader("data").load_data()
index=VectorStoreIndex.from_documents(docs)
query=index.as_query_engine()
print(query.query("Summarize the document"))
```
Sample Output:
Summary generated from the indexed document.
